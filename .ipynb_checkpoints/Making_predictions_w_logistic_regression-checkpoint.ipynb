```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import getpass
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
password = getpass.getpass()
```

```python
#conecting to the query created in mysql
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
query = '''
select title, length, rating, avg(amount) as avg_price, count(rental_date) as total_rented,max(rental_date), name from sakila.payment a
join rental b on a.rental_id = b.rental_id
join inventory c on b.inventory_id = c.inventory_id
join film d on c.film_id = d.film_id
join film_category e on d.film_id = e.film_id
join category f on e.category_id = f.category_id
group by title
order by rental_date desc;
'''

data = pd.read_sql_query(query, engine)
data
```

```python
data.dtypes
```

```python
data.describe()
```

```python
#checking nah values
data.isna().sum()
```

```python
#extracting date and month
data["rental_month"] = pd.DatetimeIndex(data["max(rental_date)"]).month
data["rental_year"] = pd.DatetimeIndex(data["max(rental_date)"]).year
data
```

```python
#input the month you wish to predict
#i am predicting month 7 since all movies rented from 8 and 7 were the exact same and the model would not be predicting anything
#this part was checked on MySQL
target_month = 7
target_year = 2006
```

```python
def data_to_analyse(x):
    if (x['rental_month']<target_month) & (x['rental_year'] <=target_year):
        return 0
    else:
        return 1
```

```python
data["target"] = data.apply(data_to_analyse, axis=1)
```

```python
data["target"].value_counts()
```

```python
data01 = data.copy()
```

```python
data01 = data01.drop(["rental_year","max(rental_date)","rental_month"], axis=1)
data01
```

```python
corr_matrix=data01.corr(method='pearson') 
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
for col in data01.select_dtypes(np.number):
    sns.displot(data01[col])
    plt.show()
```

```python
y = data01['target']
X = data01.drop('target', axis=1)
```

```python
X = pd.get_dummies(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
classification = LogisticRegression(random_state=42, multi_class='multinomial', max_iter=1000) # max_iter #multiclass multinomial é o problema
classification.fit(X_train, y_train)
```

```python
classification.score(X_test, y_test)
```

```python
predictions = classification.predict(X_test)

pd.Series(predictions).value_counts()
```

```python
y_test.value_counts()
```

```python
confusion_matrix(y_test, predictions)
```

```python
cf_matrix = confusion_matrix(y_test, predictions)
y_true = y_test
y_pred = predictions
confusion_matrix(y_true, y_pred)
```

```python
group_names = ['True A', 'False A', 'False B', 'True B']
group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
```

The model predicted that all the movies were going to be rented due to the fact of the exiting data imbalance where there have been 790 rented in the predicted month and 168 on a previous month, resulting in the model just predicting the most prevalent case and still getting 80% accuracy. A way to try to combat this problem would be by scalling the data in order for it to be balanced for the training part of the model and then apply it to the original data as the testing.
